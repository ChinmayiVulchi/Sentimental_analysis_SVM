```python


x = df['Review'].values
y = df['Liked'].values
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python



from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words = 'english')
x_train_vect = vect.fit_transform(x_train)
x_test_vect = vect.transform(x_test)
```

```python

#Apply the ML algorithm (Support vector machine)(SVM)
from sklearn.svm import SVC
model1 = SVC()
```

```python

model1.fit(x_train_vect,y_train)
```

```python


y_pred1 = model1.predict(x_test_vect)
y_pred1
```

```python

y_test
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_pred1,y_test)
```

```python

#to test the output
test = vect.transform([df['Review'][456]])
model1.predict(test)
```

```python
# pipeline method
# combines two estimators (countvect+svc)
from sklearn.pipeline import make_pipeline
model2 = make_pipeline(CountVectorizer(),SVC())

model2.fit(x_train,y_train)

y_pred2 = model2.predict(x_test)
y_pred2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred2,y_test)
```

```python


# ACCURACY FOR SVC -72%
# SVC PIPELINE - 79.2%
model2.predict([df['Review'][456]])
```

```python
#USING NAIVE BAYES
from sklearn.naive_bayes import MultinomialNB
model3 =MultinomialNB()
```

```python

model3.fit(x_train_vect,y_train)
```

```python

y_pred3 = model3.predict(x_test_vect)
y_pred3
```

```python

y_test
```

```python

accuracy_score(y_pred3,y_test)
```

```python

# to evaluate a statement and see if its spam or not using the method3
test = vect.transform([df['Review'][456]])
model3.predict(test)
```

```python
# method 4 using pipeline(countvect,multinomialNB)
from sklearn.pipeline import make_pipeline
model4 = make_pipeline(CountVectorizer(),MultinomialNB())
model4.fit(x_train,y_train)
y_pred4 = model4.predict(x_test)
y_pred4
```

```python

accuracy_score(y_pred4,y_test)
```
